```python
# Inventario inicial:
inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "11"},
    {"producto": "pantalon", "precio": 39900, "stock": "23"},
]
```

### AÑADE UN NUEVO PRODUCTO AL INVENTARIO


```python
inventario.append({"producto": "abrigo", "precio": 50000, "stock": "2"}) # se utiliza append para poder agregar 

inventario
```

### CONVIERTE LOS VALORES DEL STOCK A ENTEROS

```python
for producto in inventario: # tiene que iterar para mirar en todo el diccionario, mas especifico en el apartado de stock cambiar el tipo de variable 
    producto["stock"] = int(producto["stock"])

inventario
```

### Actualiza el precio de cada producto sumando 10.000 al precio original 

```python
for producto in inventario: # mismo bucle del item pasado, pero al tipo de variable es el mismo, por ende solo toca sumar el valor solicitado
    producto["precio"] = producto["precio"]+10000

inventario 
```

### Imprimir cada producto del inventario y sus caracteristicas 

```python
for item in inventario:
   print(f"Hay {item['stock']} unidades del producto {item['producto']}. Su precio por unidad es de {item['precio']}")
```
