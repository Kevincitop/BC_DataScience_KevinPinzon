```python
try: 
    contra = input("Ingrese contraseña: ")
    #Tengo que validar mis condiciones 
    # Condicion 1: Minimo 8 caracteres 
    if len(contra) < 8:
        raise ValueError(" --> Por favor genere una contraseña con minimo 8 caracteres <--")
    
    # Condicion 2: Al menos un digito
    digito = False # inicializo la variable
    for caracter in contra:
        if caracter.isdigit():
            digito = True
            break
    if not digito:
        raise ValueError(" --> Por favor la contraseña debe tener al menos un digito <-- ")
    
    # Condicion 3: Rectificar si contiene mayuscula 
    mayus = False 
    for caracter in contra:
        if caracter.isupper():
            mayus = True 
            break 
    if not mayus:
        raise ValueError(" --> Su contraseña debe contener al menos una mayuscula <-- ")

    print(" clave valida ")
    
except Exception as e: 
    print(e)
```
