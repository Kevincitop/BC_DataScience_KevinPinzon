**Crea un programa que permita interactuar con el usuario para recolectar datos. Asegurate que los datos recopilados tengan el formato correcto para poder hacer cálculos.**

## Estructuras de Datos (Listas, Tuplas, Diccionarios)
**Crea un programa para organizar información de múltiples ventas utilizando las estructuras adecuadas según la naturaleza de los datos.**

- **Tupla:** Crea una tupla con las categorías de productos permitidas en la tienda (ej. "Electrónica", "Ropa", "Alimentos"). Se usa una tupla porque las categorías de la tienda no cambian todos los días.
- **Diccionario:** Crea un diccionario que represente una venta específica. Debe tener las llaves: producto, categoria, precio, cantidad y total.
- **Lista:** Crea una lista vacía llamada registro_ventas y añade el diccionario de la venta que acabas de crear.

```python
# creacion de la tupla
cat_prod = ('Electronica','Ropa','Alimentos')

# Creacion de la lista 
registro_Ventas = []

# le pido al usuario que ingrese cuantas ventas desea ingresar
num_ventas = int(input(" Cuantas ventas desea registrar ")) # al optar que el numero de ventas dependede quien registre por eso genere esa potestad

for i in range(1,num_ventas + 1): # genero un bucle en 1 porque 0 ventas no van a ver y coloco un numero de ventas + 1 debido a que el rango no agarra el ultimo valor del parametro
    print(f" ____ VENTA #{i} _______")
    # pido al registrador que ingrese todos los valores que el diccionario necesita
    producto = input(" Nombre del producto : ")
    categoria = input("Categoria = Electronica, Ropa ,Alimentos")
    precio = float(input("Precio del producto"))
    cantidad = int(input("Cantidad : "))
    total = precio * cantidad
    # Genero el dicconario con todas las categorias solicitadas
    venta = { 
        "Producto" :producto,
        "Categoria" :categoria,
        "Precio" :precio,
        "Cantidad" :cantidad,
        "total" : total
    }
    # lleno la lista que inicio vacio con cada espacio del diccionario
    registro_Ventas.append(venta)
    print(f" Venta registrada. Total: ${total}") # esta impresion me garantiza que se guardo de forma correcta

print(f"\n{registro_Ventas}")# Garantizo que la lista ya no este vacia y llene con los valores ingresados por teclado

# impresion de cada parte del diccionario, por ello un ciclo, para que pueda visualizar todas las ventas 
print("\n===== REGISTRO DE VENTAS =====")
for i, venta in enumerate(registro_Ventas, 1): # Necesito que al momento de imprimir sepa a que venta se refiere, por eso lo enumero, cumpliendo con dos parametros, el primero es la variable y el ultimo es donde inicializa
    print(f"\nVenta #{i}:") # Aqui me muestra el numero de la venta 
    for clave, valor in venta.items(): # al ser mas de una clave, necesito un ciclo, utilizo el .items() para acceder a todas las claves de manera simultanea
        print(f"  {clave}: {valor}") # Para finalizar me imprime las claves 
```

## Estructuras de Control (Condicionales y Ciclos)
**Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).**

- Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
- Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
- Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).


```python
# en vez de manejar valores sueltos, ingreso el valor a una variable y asi poder manejarla de manera mas comoda 
descuento = 100000 
porcentaje = 0.10

# genero un ciclo enumerado, al contar con la lista con mas de un elementos
for i, venta in enumerate(registro_Ventas, 1): # inicializo de la misma forma, ya que las ventas empiezan a contar desde el No. 1 y no desde el cero   
    total_original = venta["total"] # creo una variable donde se guardare la posicion que voy a calcular, en este caso dentro de la lista, sacare la clave total

    # entro al condicional el cual tiene el parametro impuesto por el enunciado
    if total_original > descuento:
        total_final = total_original - (total_original * porcentaje) # calculo el descuento de forma normal
        estado_final = f"Producto con descuento del 10% : ${total_final} COP"# guardo todo dentro de una variable para poder imprimir de manera mas facil
    else:
        total_final = total_original # al no cumplir la regla, la varible no superpone el valor, si no que lo maneja de la misma forma con la que empezo
        estado_final = f"Producto sin descuento del 10% por no superar el monto de: ${descuento} COP"

    # proceso de impresion
    print(f"\nVenta #{i}")
    print(f"  Producto : {venta['Producto']}")
    print(f"  Total sin descuento : ${total_original:}")
    print(f"  {estado_final}")
    
```

## Funciones y Manejo de Errores

**Crea código reutilizable y a prueba de errores de usuario siguiendo estos parámetros.**
- Crea una función llamada calcular_ingreso que reciba dos parámetros: precio y cantidad.
- Dentro de la función, utiliza un bloque try-except. Intenta multiplicar el precio por la cantidad.
- Si ocurre un error (por ejemplo, si en lugar de un número llega un texto), el except debe atraparlo, imprimir "Error: Los valores deben ser numéricos" y retornar 0.
- Prueba tu función enviando valores correctos y luego enviando un texto para ver cómo se maneja el error.

```python
# creo la funcion con las dos variables iniciales 
def calcular_ingreso(precio, cantidad):
    # creo un manejo de errores Try para poder identificar los posibles errores que pueden dar
    try:
        if isinstance(precio, str) or isinstance(cantidad, str): # se tiene que corroborar que el tipo de variable sea verdadero, por ello se emplea el formato isinstance
            #los parametros que cumple, funciona de la siguiente manera isintance(variable,tipo) y devuelve un valor booleano, si el valor es verdadero corre a la siguiente linea, si no, salta al siguiente condicional
            raise TypeError # cumple el error 
        elif isinstance(precio, float) or isinstance(cantidad, float): # de misma forma manejamos este error
            raise ValueError
        else: # al ya superar los posibles errores, significa que los valores son correctos y que se puede hacer la operacion sin ningun problema
            print(" VALOR TOTAL ")
            print(f"$ { int(precio) * int(cantidad) } .") 

    # manejo de errores, dependiendo del llamado la impresion del error cambia 
    except TypeError:
        print(f" Error : el valor administrado no es correcto revise el precio. Los valores deben ser numericos")
        return 0

    except ValueError:
        print("Error: No se permiten numeros decimales")
        return 0
    
calcular_ingreso(20000,2)
```
