### Misión 2: Operaciones Vectorizadas
El dataset cuenta con las columnas `Precio_Unitario` y `Unidades_Vendidas`.
1. Crea una nueva columna llamada `Ingreso_Total` calculada a partir de la multiplicación de ambas columnas.
2. Muestra las primeras 5 filas para comprobar que la nueva columna se generó correctamente.

```python
df_ventas['Ingreso_Total'] = df_ventas['Precio_Unitario'] * df_ventas['Unidades_Vendidas']
df_ventas.head(5)

```

### Misión 3: Los Filtros 
La gerencia comercial necesita extraer un segmento específico.
1. Filtra el dataframe para obtener únicamente las filas donde la `Region` sea igual a `'Europa'` **Y** el `Ingreso_Total` sea mayor a `500`.
2. Guarda este resultado en una nueva variable llamada `df_europa_top` y muéstrala en pantalla.

```python
columnas_interes = ['Videojuego', 'Plataforma', 'Region','Ingreso_Total']

resultado = df_ventas[
    (df_ventas['Region'] == 'Europa') &
    (df_ventas['Ingreso_Total'] > 500)
]

resultado[columnas_interes].head()
```

### Misión 4: Agrupación y Respuestas de Negocio
Queremos evaluar el rendimiento por consola de videojuegos.
1. Usando el dataframe original (`df_ventas`), agrupa los datos por la columna `Plataforma`.
2. Suma las `Unidades_Vendidas` por cada una.
3. Almacena este resumen en una variable llamada `reporte_plataformas` y muéstralo en pantalla.
4. Responde en una celda de texto: ¿Cuál es la plataforma o consola con más unidades vendidas en este conjunto de datos?

```python
reporte_plataformas = df_ventas.groupby('Plataforma')['Unidades_Vendidas'].sum()

print(f" La consola mas vendida es :", reporte_plataformas.idxmax(), "y con un numero de unidades :", reporte_plataformas.max())
reporte_plataformas
```
